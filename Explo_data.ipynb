```python
# plot first few images
plt.figure(figsize = (14, 14))
for i in range(9):
    try:
        # define subplot
        plt.subplot(330 + 1 + i)
        # define filename
        filename = df['image'][i]
        # load image pixels
        image = imread('../data/Flipkart/Images/' + filename)
        # plot raw pixel data
        plt.imshow(image)
    except:
        print(f'Erreur : no such file or directory: {filename}')
## show the figure
plt.show()
```

```python
# plot 25 images au hasard
plt.figure(figsize = (15, 11))

for i in range(0, 25):
    plt.subplot(5, 5, 1 + i)
    x = int(np.random.randint(0, 1050, 1))
    image = imread('../data/Flipkart/Images/' + df['image'][x])
    plt.title(f'#{x}')
    plt.imshow(image)

plt.tight_layout()
plt.show()
```
